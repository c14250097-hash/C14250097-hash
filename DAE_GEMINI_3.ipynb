```python
import pandas as pd
from google.colab import files

# Langkah 1: Unggah file dari komputer Anda
uploaded = files.upload()

# Langkah 2: Dapatkan nama file yang diunggah
file_name = next(iter(uploaded))

# Langkah 3: Baca file menggunakan Pandas
df = pd.read_csv(file_name)

# Langkah 4: Tampilkan data untuk verifikasi
print(df.head())

import matplotlib.pyplot as plt

# Buat histogram
plt.hist(df['Observed Length (m)'], bins=20, color='skyblue', edgecolor='black')

# Tambahkan judul dan label
plt.title('Distribusi Panjang Buaya')
plt.xlabel('Panjang (m)')
plt.ylabel('Frekuensi')

# Tampilkan plot
plt.show()
```
